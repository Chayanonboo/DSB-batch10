```python
## Create import request
import requests
from requests import get
# !pip install gazpacho
!pip install gazpacho
from gazpacho import Soup
from requests import get
```

```python
url = "https://www.verywellmind.com/how-to-become-a-more-effective-learner-2795162"
```

```python
response = get("https://www.verywellmind.com/how-to-become-a-more-effective-learner-2795162")

response.status_code
```

```python
response.text
```

```python
web = get(url)

how_to_learn = Soup(web.text)

print(type(how_to_learn))
```

```python
## find information from this Soup
how_to_learn.find("h2", mode = "first").strip()
```

```python
# extract website title
how_to_learn.find("title").strip()
```

```python
## find heading 2 from this soup
for h2 in how_to_learn.find("h2"):
    print(h2.strip())
```

```python
## find heading 3 from this soup
for h3 in how_to_learn.find("h3"):
    print(h3.strip())
```

```python
# extract paragraphs
for paragraph in how_to_learn.find("p"):
  print(paragraph.text)
```

```python
# extract links
for link in how_to_learn.find("a"):
    print(link.strip())
```
