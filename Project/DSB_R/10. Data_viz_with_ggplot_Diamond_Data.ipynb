**sample fraction of diamonds dataset**

```r
set.seed(27)
df <- diamonds %>%
sample_frac(0.04)
```

**graph 1** to show scatter plot relation of carat and price variable

```r
ggplot(df, aes(carat, price, colour = cut)) +
  geom_point(alpha = 0.9) +
  geom_rug() +
  theme_minimal() +
  labs(title = "Relation between Price and Diamond Carat",
  x = "Diamond Carat",
  y = "Price")
```

graph 2 to show scatter plot relation of depth and table variable

```r
ggplot(df, aes(colour = cut, clarity, table)) +
  geom_point(alpha = 0.5) +
  geom_point() +
  theme_minimal() +
  labs(title = "Relation between depth and Diamond table",
  x = "Depth",
  y = "Table")
```

**graph 3 bar chart show distribution of y**

```r
ggplot(df, aes(y)) +
      geom_histogram(fill = "red") +
      theme_minimal() +
      labs(title = "Distribution of y",
            x= "No. of Diamonds",
            y= "y")
```

graph 4 show scatter plot with lm model with relation between carat and price

```r
ggplot(df, aes(carat,price)) +
  geom_point(alpha=0.3, size=2, color="red") +
  geom_smooth(method="lm", color="blue") +
  theme_minimal() +
  facet_grid(~ cut)
```

**graph 5 show scatter plot of depth vs. table by color**


```r
ggplot(df, aes(x = table, y = depth, color = color)) +
  geom_point(alpha = 0.9) +
  facet_wrap(~ color, ncol = 2) +
  theme_minimal() +
  labs(
    title = "Scatter Plot of Depth vs. Table by Color",
    x = "Table",
    y = "Depth"
  )
```
