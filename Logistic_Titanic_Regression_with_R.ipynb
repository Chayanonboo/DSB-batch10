```r
install.packages("titanic")
library(titanic)

View(Titanic)
head(titanic_train)

## survived 0 life 1 die
## p class

## DROP NA (missing values)
titanic_train <- na.omit(titanic_train)
nrow(titanic_train)

## SPLIT DATA
set.seed(2)
n <- nrow(titanic_train)
id <- sample(1:n, size=n*0.7) ## 70% train 30% test
train_data <- titanic_train[id, ]
test_data <- titanic_train[-id, ]

nrow(train_data)
nrow(test_data)

## train Model
model <- glm(Survived~Pclass+Sex+Age+SibSp, data = train_data, family="binomial")
summary(model)
train_data$prob_survived <- predict(model, type="response")
train_data$pred_survived <- ifelse(train_data$prob_survived >= 0.5,1,0)

## View data
train_data$prob_survived
train_data$pred_survived

## confusion matrix
conM_train <- table(train_data$pred_survived,train_data$Survived,
                    dnn=c("Prediced","Actual"))

## Model Evaluation
conM_train[1,1]
conM_train[1,2]
conM_train[2,1]
conM_train[2,2]

## Accuracy
cat("accuracy:",(conM_train[1,1] + conM_train[2,2]) / sum(conM_train))
cat("Precision:",conM_train[2,2] / (conM_train[2,1] + conM_train[2,2]))
cat("Recall:",conM_train[2,2] / (conM_train[1,2] + conM_train[2,2]))
cat("F1:",2 * ((0.9*0.9) / (0.9+0.9)))

## Fit Logistic Regression test model Data Set
test_model <- glm(Survived~Pclass+Sex+Age+SibSp, data = test_data, family="binomial")
summary(test_model)

## test model
test_data$prob_survived <- predict(test_model, type="response")
test_data$pred_survived <- ifelse(test_data$prob_survive>=0.5,1,0)

conM_test <- table(test_data$pred_survive,test_data$Survived,
                   dnn=c("Prediced","Actual"))

## Model Evaluation test model
conM_test[1,1]
conM_test[1,2]
conM_test[2,1]
conM_test[2,2]

## Accuracy
cat("accuracy:",(conM_test[1,1] + conM_test[2,2]) / sum(conM_test))
cat("Precision:",conM_test[2,2] / (conM_test[2,1] + conM_test[2,2]))
cat("Recall:",conM_test[2,2] / (conM_test[1,2] + conM_test[2,2]))
cat("F1:",2 * ((0.9*0.9) / (0.9+0.9)))

## Compare Accuracy train data vs test data
cat("Train data \n Accuracy:",(conM_train[1,1] + conM_train[2,2]) / sum(conM_train), "\n")

cat("Test data \n Accuracy:",(conM_test[1,1] + conM_test[2,2]) / sum(conM_test), "\n")
```
